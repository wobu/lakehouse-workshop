## DDL

```python
%%sql
CREATE SCHEMA IF NOT EXISTS demo.nyc;
```

```python
%%sql
CREATE TABLE demo.nyc.taxis
(
  vendor_id bigint,
  trip_id bigint,
  trip_distance float,
  fare_amount double,
  store_and_fwd_flag string
)
PARTITIONED BY (vendor_id);
```

## DML
